```python
import streamlit as st
import pandas as pd 

# 1. Initialize the connection object
# 'mysql' should match the section name in your secrets.toml file ([connections.mysql])
conn = st.connection('mysql', type='sql')

# 2. Define a function to query the database and cache the results
# @st.cache_data is important to prevent re-running the query on every Streamlit rerun
@st.cache_data(ttl=600) # ttl=600 means data is cached for 10 minutes (600 seconds)
def run_query(query):
    # conn.query runs the SQL query and returns the results as a pandas DataFrame
    return conn.query(query, ttl=600)

# 3. Use the function to fetch data
# Replace 'your_table_name' with the actual table you want to query
df = run_query("SELECT * FROM policelog_data;")

# 4. Display the results in your Streamlit app
st.title("Streamlit & MySQL Data Display")

st.subheader("Data from MySQL")
st.dataframe(df)

# You can iterate over the DataFrame or display specific information
for row in df.itertuples():
    st.write(f"Record: ID {row.Index + 1}, Column_A: {row.column_a}")
```
